```python
df_normal = df_normal.drop(columns='class', axis=0)
df_tcp_syn = df_tcp_syn.drop(columns='class', axis=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Comparação das médias das colunas
mean_normal = df_normal.mean()
mean_tcp_syn = df_tcp_syn.mean()

# Criar um dataframe para facilitar a visualização
mean_df = pd.DataFrame({
    'normal': mean_normal,
    'tcp_syn': mean_tcp_syn
})

# Plotar as médias
mean_df.plot(kind='bar', figsize=(15, 7))
plt.title('Comparação das Médias das Métricas SNMP')
plt.ylabel('Valor Médio')
plt.show()

# Boxplots para comparar a distribuição das colunas
for column in df_normal.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=[df_normal[column], df_tcp_syn[column]], palette='Set3')
    plt.xticks([0, 1], ['Normal', 'TCP SYN'])
    plt.title(f'Comparação da Distribuição de {column}')
    plt.show()

```

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Tabelas para armazenar os resultados
t_test_results = {}
mann_whitney_results = {}

for column in df_normal.columns:
    t_test_results[column] = ttest_ind(df_normal[column], df_tcp_syn[column])
    mann_whitney_results[column] = mannwhitneyu(df_normal[column], df_tcp_syn[column])

# Exibir resultados
print("Resultados do Teste T de Student:")
for column, result in t_test_results.items():
    print(f"{column}: statistic={result.statistic}, p-value={result.pvalue}")

print("\nResultados do Teste de Mann-Whitney:")
for column, result in mann_whitney_results.items():
    print(f"{column}: statistic={result.statistic}, p-value={result.pvalue}")
```

```python
import matplotlib.pyplot as plt

# Lista de colunas da interface e TCP com maiores desvios
significant_interface_metrics = [
    'ifOutOctets11', 'ifInDiscards11'
]
significant_tcp_metrics = [
    'tcp?ActiveOpens', 'tcpEstabResets', 
    'tcpCurrEstab']

# Função para plotar comparações
def plot_comparison(metrics, df_normal, df_tcp_syn, title_prefix):
    for column in metrics:
        plt.figure(figsize=(10, 5))
        plt.plot(df_normal[column], label='Normal', marker='o')
        plt.plot(df_tcp_syn[column], label='TCP SYN', marker='x')
        plt.title(f'{title_prefix}: {column}')
        plt.xlabel('Amostra')
        plt.ylabel(column)
        plt.legend()
        plt.show()

# Plotar comparações das métricas de interface
plot_comparison(significant_interface_metrics, df_normal, df_tcp_syn, 'Interface Metrics')

# Plotar comparações das métricas de TCP
plot_comparison(significant_tcp_metrics, df_normal, df_tcp_syn, 'TCP Metrics')
```
